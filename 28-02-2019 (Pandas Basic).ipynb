```python
### Creating a random normal distinbusion from 1 to 4

import numpy as np
np.random.seed(2)
norm_dist = pd.DataFrame(np.random.randn(1,4))
norm_dist
```

```python
### Creating a binomial distinbusion

binom_dist = pd.DataFrame(np.random.binomial(100,0.5,(1,4)))
binom_dist
```

```python
### Creating a poisson distinbusion

poisson_dist = pd.DataFrame(np.random.poisson(100,(1,4)))
poisson_dist
```

```python
dist = [norm_dist,binom_dist,poisson_dist]
dist
```

```python
df5 = pd.concat(dist, keys=['Normal','Binomal','Poisson'])
df5
```

```python
## Slice the first 2 rows & 3 cols

s1 = df5.iloc[:2,:3]
s1
```

```python
## Slice the remaining rows

s2 = df5[2:]
s2
```

```python
pd.concat([s1,s2],axis=0)
```
