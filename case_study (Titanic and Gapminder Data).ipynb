```python
life_expectancy = pd.read_csv("D:/programming in home/Case_Study/life_expectancy.csv")
regions = pd.read_csv("D:/programming in home/Case_Study/regions.csv")
```

```python
gapminder = pd.read_csv("D:/programming in home/Case_Study/gapminder.csv")
gapminder.head()
```

```python
df = gapminder.set_index(['Year','region','Country'])
df.head()
```

```python
def spread(ser):
    return (max(ser) - min(ser))
aggregator ={'population':'sum', 'child_mortality':'mean' , 'gdp':spread }
df1 = df.groupby(level=['Year','region']).agg(aggregator).reset_index()
df1.tail()
```
