```python
import matplotlib.pyplot as plt

def walk(d,s):     ## function walk run to set the step 
    if d <=2  :
        s = max(0,s-1)
    elif d <= 5:
        s +=1
    else:
        d = np.random.randint(1,7)
        s = walk(d,s)
    return s

all_walk = []
for j in range(10):
    random_walk = [0]
    for i in range(500):
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        step = walk(dice,step)
        
        if np.random.rand() <= 0.001 :
            step = 0
            
        random_walk.append(step)
        
    all_walk.append(random_walk)

aw = np.array(all_walk)
aw_t = np.transpose(aw)
end = np.array(aw_t[-1])

graph = plt.plot(aw_t)
plt.show(graph)

histo = plt.hist(end)
plt.show(histo)
```
